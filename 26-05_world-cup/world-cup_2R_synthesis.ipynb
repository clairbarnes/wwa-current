# Synthesise new estimates

```r
res_obs <- read.csv("res-obs_vrp.csv", row.names = "X")
res_mdl <- read.csv("res-mdl_vrp.csv", row.names = "X")

# correct column names for consistency
colnames(res_obs) <- gsub("_","",colnames(res_obs))
colnames(res_mdl) <- gsub("PR","pr",colnames(res_mdl))
```

```r
for (stadium in rownames(res_obs)) {
    df_obs <- res_obs[stadium,]
    rownames(df_obs) <- "ERA5-Land"
    
    df_mdl <- res_mdl[gsub("_.+","",rownames(res_mdl)) == stadium,]
    rownames(df_mdl) <- gsub(".+_","",rownames(df_mdl))

    for (cnm in c("dI_abs", "pr26", "pr28", "pr30", "pr32")) {
        if (cnm == "dI_abs") stype <- "abs" else stype <- "PR"

        for (period in c("attr", "proj")) {

            synth_fnm <- paste0("synth_vrp/synth-vrp_",tolower(gsub("_.+","",cnm)), "-",period, "_", stadium,".csv")
            print(synth_fnm)
            
            if (period == "attr") {
                obs_in <- df_obs[,grepl(gsub("_", "", cnm), colnames(df_obs))][,c(1,2,4)]
            } else {
                obs_in <- NA
            }

            mdl_in <- df_mdl[,grepl(paste0(period,"_", gsub("_",".",cnm)), colnames(df_mdl))][,c(1,2,4)]
            mdl_in <- mdl_in[rowSums(is.na(mdl_in)) == 0,]
                       
            if (nrow(mdl_in) == 0) next

            if(all(!is.finite(mdl_in[,1]))) {
                # if (period == "attr") {
                #     synth <- rbind(setNames(obs_in, c("est", "lower", "upper")), setNames(mdl_in, c("est", "lower", "upper")))
                # } else {
                #     synth <- setNames(mdl_in, c("est", "lower", "upper"))
                # }
                # write.csv(synth, synth_fnm)
                next
            }

            if (stype == "PR") {
                if (period == "attr") obs_in <- infer_infinite(obs_in)
                mdl_in <- infer_infinite(mdl_in)
            }
    
            synth <- synthesis(obs_in, mdl_in, stype)$df
            write.csv(synth, synth_fnm, row.names = F)
        }
    }
}
```
