```python
import pandas as pd, matplotlib, matplotlib.pyplot as plt, glob, numpy as np, re
from matplotlib import collections as mc
matplotlib.rcParams['savefig.bbox'] = "tight"    # always save with tight bounding box

stadia = sorted(["Atlanta", "Boston", "Dallas", "Guadalajara", "Houston", "Los-Angeles", "Mexico-City", "Monterrey", "Toronto", "Vancouver", "Miami", "Seattle", "Kansas", "San-Francisco", "New-York", "Philadelphia"])
```

```python
period = "attr"  
for stype in ["di", "pr26", "pr28", "pr30", "pr32"]:

    fl = sorted(glob.glob("synth_vrp/synth-vrp_"+stype+"-"+period+"_*"))
    nc = 4; nr = 4
    fig, axs = plt.subplots(ncols = nc, nrows = nr, figsize = (nc*3.5, nr * 3), dpi = 100, sharex = True, sharey = True)
    plt.subplots_adjust(wspace = 0.1)
    
    model_lookup = {"CNRM-CM6-1-HR": 0, "GFDL-CM4": 1, "HadGEM3-GC31-MM" : 2, "INM-CM5-0":3, "NorESM2-MM": 4}
    markerprops = {"marker" : "|", "ms" : 8, "mew" : 2}
    lineprops = {"linewidths" : 6}
    
    for i in range(len(stadia)):
        
        ax = axs.flatten()[i]
        stadium = stadia[i]
        fnm_check = [fnm.split("_")[-1][:-4] == stadium for fnm in fl]
    
        ax.set_title(re.sub("-|_", " ", stadium), fontweight = "bold")
    
        if sum(fnm_check) == 0: 
            continue
    
        fnm = [fnm for fnm in fl if fnm.split("_")[-1][:-4] == stadium][0]
        s_df = pd.read_csv(fnm, index_col = "group")
        
        df_obs = s_df.loc[["obs"]]
        df_models = s_df.loc["models"]
        df_modelsynth = s_df.loc[["model_synth"]]
        df_synth = s_df.loc[["synth"]]
        
        x = 8
        ax.add_collection(mc.LineCollection([[(df_obs.iloc[i].lower,x),(df_obs.iloc[i].upper,x)] for i in range(len(df_obs))], 
                                            colors = "blue", alpha = 1/len(df_obs), **lineprops))
        [ax.plot(df_obs.iloc[i].est, x, marker = "d", mec = "k", mfc = "blue") for i in range(len(df_obs))]
    
        # loop over model names and plot at corresponding y to make sure models line up
        x = x-1.5
        for m, i in model_lookup.items():
            if not m in df_models.model.values: continue
            r = df_models.loc[df_models.model == m]
            ax.add_collection(mc.LineCollection([[(r.iloc[j].lower,x-i),(r.upper.iloc[j],x-i)] for j in range(len(r))], colors = "red", alpha = 0.6, **lineprops))
            ax.add_collection(mc.LineCollection([[(r.iloc[j].l_wb,x-i),(r.u_wb.iloc[j],x-i)] for j in range(len(r))], colors = "red", alpha = 0.1, **lineprops))       
            ax.plot(r.est, x-i, mec = "red", mfc = "red", alpha = 0.5, **markerprops)
    
        x = x - 5
        ax.add_collection(mc.LineCollection([[(df_modelsynth.iloc[i].lower,x),(df_modelsynth.iloc[i].upper,x)] for i in range(len(df_modelsynth))], 
                                            colors = "red", alpha = 1, **lineprops))
        ax.plot(df_modelsynth.est, x, marker = "d", mec = "k", mfc = "red")
            
        x = x-1.5
        ax.add_collection(mc.LineCollection([[(df_synth.iloc[i].lower,x),(df_synth.iloc[i].upper,x)] for i in range(len(df_synth))],
                                            colors = "magenta", alpha = 1, **lineprops))
        ax.plot(df_synth.est, x, marker = "d", mec = "k", mfc = "magenta")
    
        if stype.lower() == "di":
            ax.axvline(0, color = "k", ls = "--")
            ax.set_xlim(-1.5,3)
            fig.text(0.5, 0.065, "Change in intensity (°C)", ha='center', va = "center", fontsize = "large")
            
        else:
            ax.set_xscale('log')
            ax.axvline(1, color = "k", ls = "--")
            ax.set_xlim(1e-1,1e3)
            fig.text(0.5, 0.07, "Probability ratio", ha='center', va = "center", fontsize = "large")
    
        ax.set_yticks([8] + list(np.arange(6.5,0.5,-1)) + [0])
        ax.set_yticklabels(["ERA5-Land"] + list(model_lookup.keys()) + ["Model synthesis", "Synthesis"])
        ax.grid(color = "k", alpha = 0.1, axis = "x")
    
    plt.savefig("fig_world-cup_"+stype+"-"+period); plt.close()
```
