# **Distributions of maxima**

- **_What sort of shape parameters should we expect to see for GEV fitted to maxima of different distributions?_**

Raised by an observation from Steve Jewson: typically, people use normal distributions for modelling daily temperature; according to the theory, the maxima of normal distributions should be gumbel distributed. However, we often find a negative shape parameter for temperature maxima.

```r
suppressMessages(library("extRemes"))

prep_window <- function(rc = c(1,1), w = 4, h = 4, ...) { options(repr.plot.width = rc[2]*w, repr.plot.height = rc[1]*h, repr.plot.res = 200); par(mfrow = rc, pch = 20, ...) }

panel.cor <- function(x, y, digits=2, prefix="r = ", cex.cor) {
    usr <- par("usr"); on.exit(par(usr))
    par(usr = c(0, 1, 0, 1))
    r <- cor(x, y)
    txt <- format(c(r, 0.123456789), digits=digits)[1]
    txt <- paste(prefix, txt, sep="")
    if(missing(cex.cor)) cex <- 0.8/strwidth(txt)
    text(0.5, 0.5, txt, cex = cex * abs(r))
}
```

## **X ~ N**
- Expect Gumbel distribution, shape close to zero

```r
set.seed(42)
nmax <- sapply(1:10000, function(s) max(rnorm(1000)))  
nmax_pars <- fevd(nmax)$results$par
nmax_shape <- round(nmax_pars["shape"], 2)

prep_window(c(1,3)); par(mfrow = c(1,2))

x <- seq(-5,5,0.1)
plot(x, dnorm(x), type = "l", xlab = "", ylab = "Density", main = "X ~ N(0,1)")
# boxplot(nmax, horizontal = T, at = 0.3, add = T, boxwex = .3)

x <- seq(2,6,0.1)
plot(density(nmax, from = min(x), to = max(x)), lwd = 2, main = bquote("X normal, "*xi==.(nmax_shape)), xlab = "")
lines(x, devd(x, loc = nmax_pars["location"], scale = nmax_pars["scale"], shape = nmax_pars["shape"]), col = "blue", lwd = 2)
```

## **X ~ Student-t (heavy tails)**
- heavy tails so expect Frechet distribution with negative shape parameter - yet we get a positive shape parameter here somehow!


```r
set.seed(42)
df_list <- c(20,10,5); df_cols <- c("blue", "cornflowerblue", "skyblue")
t_max <- sapply(df_list, function(k) sapply(1:10000, function(s) max(rt(1000, k))))
t_pars <- apply(t_max, 2, function(Y) fevd(Y)$results$par)
```

```r
prep_window(c(1,3))

x <- seq(-5,5,0.1)
plot(x, dnorm(x), type = "l", xlab = "", ylab = "Density", main = "X", lwd = 2, ylim = c(0,0.5))
for(i in 1:length(df_list)) lines(x, dt(x, df_list[i]), type = "l", col = df_cols[i], lwd = 2)
legend("topright", legend = c("Normal", paste0("t_",df_list)), lty = 1, lwd = 2, col = c("black", df_cols))

x <- seq(2,10,0.1)
plot(density(nmax, from = min(x), to = max(x)), lwd = 2, main = "max (X)", xlab = "")
for(i in 1:length(df_list)) {
    lines(density(t_max[,i], from = min(x), to = max(x)), lwd = 1, col = df_cols[i])
    lines(x, devd(x, loc = t_pars["location",i], scale = t_pars["scale",i], shape = t_pars["shape",i]), col = df_cols[i], lwd = 2)
}

plot(c(0, 6/(df_list-4)), c(nmax_pars["shape"], t_pars["shape",]), col = c("black", df_cols), pch = 4, lwd = 2,
     xlab = "Excess kurtosis of X", ylab = "Shape of max (X) GEV", main = "Shape parameter vs excess kurtosis")
abline(h = 0, lty = 2)
```

## **X ~ skewnormal**

- positive skewness (long upper tail) doesn't make much difference to distribution of maximum, hence also not to shape parameter
- when skewness is negative (long lower tail), the value of $\alpha$ has more effect on maxima & so on shape parameter
- however, effect of skewness alone is quite small: for realistic values, $\xi$ is generally between -0.05 and -0.1

```r
suppressWarnings(library(sn))
```

```r
set.seed(42)
skew_list <- c(-2,-1,-0.5,0.5,1,2); skew_cols <- c("blue", "cornflowerblue", "skyblue", "gold", "orange", "red4")

sn_max <- sapply(skew_list, function(k) sapply(1:10000, function(s) max(rsn(1000, alpha = k))))
sn_pars <- apply(sn_max, 2, function(Y) fevd(Y)$results$par)
```

```r
prep_window(c(1,4))

x <- seq(-5,5,0.1)
plot(x, dnorm(x), type = "l", xlab = "", ylab = "Density", main = "X", lwd = 2, ylim = c(0,0.7))
for(i in 1:length(skew_list)) lines(x, dsn(x, alpha = skew_list[i]), type = "l", col = skew_cols[i], lwd = 2)
legend("top", legend = c("Normal", paste0("Skewed, ",skew_list)), lty = 1, lwd = 2, col = c("black", skew_cols), ncol = 2)

x <- seq(-3,3, 0.01)
plot(abs(x - x[which.max(dnorm(x))]), dnorm(x), type = "l", xlab = "", ylab = "Density", main = "Asymmetry of X", lwd = 2, ylim = c(0,0.7))
for(i in 1:length(skew_list)) {
    y <- dsn(x, alpha = skew_list[i])
    lines(abs(x - x[which.max(y)]), y, col = skew_cols[i], lwd = 2)
}

x <- seq(0,6,0.01)
plot(density(nmax, from = min(x), to = max(x)), lwd = 2, main = "max (X)", xlab = "", ylim = c(0,2.6))
for(i in 1:length(skew_list)) {
    lines(density(sn_max[,i], from = min(x), to = max(x)), lwd = 1, col = skew_cols[i])
    lines(x, devd(x, loc = sn_pars["location",i], scale = sn_pars["scale",i], shape = sn_pars["shape",i]), col = skew_cols[i], lwd = 2)
}

plot(c(0, skew_list), c(nmax_pars["shape"], sn_pars["shape",]), col = c("black", skew_cols), pch = 4, lwd = 2, ylim = c(-0.1,0.1),
     xlab = "Skewness of X", ylab = "Shape of max (X) GEV", main = "Shape parameter vs skewness")
abline(h = 0, v = 0, lty = 2)
```

```r
# check for more skewed data
set.seed(42)
nskew_list <- c(-200, -100,-50,-25,-10,-5,-2.5); nskew_cols <- c("darkorchid4", "midnightblue", "blue", "cornflowerblue", "skyblue", "chartreuse3", "forestgreen")

nsn_max <- sapply(nskew_list, function(k) sapply(1:10000, function(s) max(rsn(1000, alpha = k))))
nsn_pars <- apply(nsn_max, 2, function(Y) suppressWarnings(fevd(Y))$results$par)
                  
delta <- nskew_list / sqrt(1+nskew_list^2)
kurtosis <- 2*(pi-3) * (delta*sqrt(2/pi))^4 / (1-2*(delta^2) / pi)^2 

prep_window(c(1,3))
plot(nskew_list, nsn_pars["shape",], col = nskew_cols, pch = 4, lwd = 2,
     xlab = "Skewness of X", ylab = "Shape of max (X) GEV", main = "Shape parameter vs skewness")

plot(kurtosis, nsn_pars["shape",], col = nskew_cols, pch = 4, lwd = 2,
     xlab = "Excess kurtosis of X", ylab = "Shape of max (X) GEV", main = "Shape parameter vs excess kurtosis")
                  
plot(x, dnorm(x), type = "n", xlab = "", ylab = "Density", main = "X", lwd = 2, ylim = c(0,0.8))
for(i in 1:length(nskew_list)) lines(x, dsn(x, alpha = nskew_list[i]), type = "l", col = nskew_cols[i], lwd = 2)
# legend("top", legend = c("Normal", paste0("Skewed, ",skew_list)), lty = 1, lwd = 2, col = c("black", skew_cols), ncol = 2)
```

## **X ~ skew-t**
- $\xi$ is more linearly related to the skewness of the observed data than the kurtosis
- Varying the skewness & degrees of freedom together makes very clear that kurtosis has a much stronger effect on the shape parameter than alpha


```r
set.seed(42)
skew_list <- c(-10,-5,-2,-1,0,5); skew_cols <- c("blue", "cornflowerblue", "skyblue", "aquamarine", "black", "gold")
df_list <- c(50,20,10,7,5); df_cols <- c("black", "blue", "cornflowerblue", "skyblue", "aquamarine")

tmax_pars <- abind::abind(sapply(skew_list, function(skew) {
    sapply(df_list, function(df) {
        samp <- sapply(1:10000, function(s) max(rst(1000, alpha = skew, nu = df)))
        c(fevd(samp)$results$par, "kurtosis" = moments::kurtosis(samp), "skew" = moments::skewness(samp), "alpha" = skew, "nu" = df)
    })
}, simplify = F), rev.along = 0)
```

```r
prep_window(c(1,3))
matplot(df_list, tmax_pars["location",,], pch = 4, lwd = 2, col = skew_cols,
        main = "GEV location parameter", xlab = "Degrees of freedom", ylab = "GEV location parameter")

matplot(df_list, tmax_pars["scale",,], pch = 4, lwd = 2, col = skew_cols,
        main = "GEV scale parameter", xlab = "Degrees of freedom", ylab = "GEV scale parameter")

matplot(df_list, tmax_pars["shape",,], pch = 4, lwd = 2, col = skew_cols, ylim = c(-0.1, 0.2),
        main = "GEV shape parameter", xlab = "Degrees of freedom", ylab = "GEV shape parameter")
abline(h = 0, lty = 2)
legend("topright", legend = skew_list, col = skew_cols, lty = 1, lwd = 2, title = "Skewness")
```

```r
prep_window(c(1,3))
matplot(skew_list, t(tmax_pars["location",,]), pch = 4, lwd = 2, col = df_cols,
        main = "GEV location parameter", xlab = "Alpha", ylab = "GEV location parameter")
legend("topleft", legend = df_list, col = df_cols, lty = 1, lwd = 2, title = "Degrees of freedom")

matplot(skew_list, t(tmax_pars["scale",,]), pch = 4, lwd = 2, col = df_cols,
        main = "GEV scale parameter", xlab = "Alpha", ylab = "GEV scale parameter")

matplot(skew_list, t(tmax_pars["shape",,]), pch = 4, lwd = 2, col = df_cols, ylim = c(-0.1, 0.2),
        main = "GEV shape parameter", xlab = "Alpha", ylab = "GEV shape parameter")
abline(h = 0, lty = 2)
```

```r
prep_window(c(1,4))#; par(cex = 1)

plot(rep(df_list, length(skew_cols)), tmax_pars["kurtosis",,], col = rep(df_cols, length(skew_cols)), pch = rep(0:(length(skew_cols)-1), each = length(df_cols)), lwd = 2, 
     xlab = "Degrees of freedom", ylab = "Sample kurtosis", main = "DF vs sample kurtosis")
legend("topright", legend = df_list, col = df_cols, lty = 1, lwd = 2, title = "Degrees of freedom", bty = "n")

plot(tmax_pars["kurtosis",,], lwd = 2, tmax_pars["shape",,], col = rep(df_cols, length(skew_cols)), xlab = "Sample kurtosis", ylab = expression(xi), pch = rep(0:(length(skew_cols)-1), each = length(df_cols)),
     main = "Sample kurtosis vs GEV shape", ylim = c(-0.1, 0.2))

plot(rep(skew_list, each = length(df_cols)), tmax_pars["skew",,], col = rep(df_cols, length(skew_cols)), pch = rep(0:(length(skew_cols)-1), each = length(df_cols)), lwd = 2, ylim = c(0,5),
     xlab = expression(alpha), ylab = "Sample skewness", main = "Shear vs sample skewness")
legend("topright", legend = skew_list, col = "black", pch = 0:(length(skew_cols)-1), lwd = 2, lty = NA, title = "Shear parameter", bty = "n")

plot(tmax_pars["skew",,], tmax_pars["shape",,], lwd = 2, col = rep(df_cols, length(skew_cols)), xlab = "Sample skewness", ylab = expression(xi), pch = rep(0:(length(skew_cols)-1), each = length(df_cols)),
     main = "Sample skewness vs GEV shape", ylim = c(-0.1, 0.2))
```

## **X ~ gamma**
- Location & scale of GEV are linearly dependent on scale of Gamma distribution & approximately linearly dependent on shape of Gamma distribution
- For Gamma shape close to 1, $\xi$ will be close to zero - otherwise $\xi$ will be negative
- For large values of the Gamma shape parameter $\xi$ decreases - however, very little effect for small changes in scale & shape

### **Effect of changing shape parameter**

```r
set.seed(42)

shape_list <- c(1,1.5,2,5); shape_cols <- c("blue", "cornflowerblue", "skyblue", "aquamarine")

g_max <- sapply(shape_list, function(k) sapply(1:10000, function(s) max(rgamma(1000, shape = k))))
g_pars <- apply(g_max, 2, function(Y) fevd(Y)$results$par)
```

```r
prep_window(c(1,4))

x <- seq(0,10,0.1)
plot(x, dnorm(x), type = "n", xlab = "", ylab = "Density", main = "X", lwd = 2, ylim = c(0,1))
for(i in 1:length(shape_list)) lines(x, dgamma(x, shape_list[i]), type = "l", col = shape_cols[i], lwd = 2)
legend("topright", legend = paste0("shape = ",shape_list), lty = 1, lwd = 2, col = shape_cols)

x <- seq(4,25,0.1)
plot(density(nmax, from = min(x), to = max(x)), type = "n", lwd = 2, main = "max (X)", xlab = "", ylim = c(0,0.4))
for(i in 1:length(shape_list)) {
    lines(density(g_max[,i], from = min(x), to = max(x)), lwd = 1, col = shape_cols[i])
    lines(x, devd(x, loc = g_pars["location",i], scale = g_pars["scale",i], shape = g_pars["shape",i]), col = shape_cols[i], lwd = 2)
}

plot(shape_list, g_pars["shape",], col = shape_cols, pch = 4, lwd = 2, ylim = c(-0.1, 0.1), 
     xlab = "Shape parameter of X", ylab = "Shape of max (X) GEV", main = "GEV shape parameter vs gamma shape parameter")
abline(h = 0, lty = 2)

plot(6/shape_list, g_pars["shape",], col = shape_cols, pch = 4, lwd = 2, ylim = c(-0.1, 0.1), 
     xlab = "Excess kurtosis of X", ylab = "Shape of max (X) GEV", main = "GEV shape parameter vs gamma excess kurtosis")
abline(h = 0, lty = 2)
```

```r
prep_window(c(1,3))

plot(shape_list, g_pars["location",], col = shape_cols, pch = 4, lwd = 2, 
     xlab = "Shape parameter of X", ylab = "Location of max (X) GEV", main = "GEV location vs gamma shape parameter")
abline(line(shape_list, g_pars["location",]), lty = 2)

plot(shape_list, g_pars["scale",], col = shape_cols, pch = 4, lwd = 2, 
     xlab = "Shape parameter of X", ylab = "Scale of max (X) GEV", main = "GEV scale vs gamma shape parameter")
abline(line(shape_list, g_pars["scale",]), lty = 2)

plot(shape_list, g_pars["shape",], col = shape_cols, pch = 4, lwd = 2, ylim = c(-0.1,0.1),
     xlab = "Shape parameter of X", ylab = "Shape of max (X) GEV", main = "GEV shape vs gamma shape parameter")
abline(line(shape_list, g_pars["shape",]), lty = 2)
```

### **Effect of changing scale parameter**

```r
set.seed(42)

scale_list <- c(1,2,3,4,5); scale_cols <- c("blue", "cornflowerblue", "skyblue", "aquamarine", "green2")

g_max <- sapply(scale_list, function(k) sapply(1:10000, function(s) max(rgamma(1000, shape = 2, scale = k))))
g_pars <- apply(g_max, 2, function(Y) fevd(Y)$results$par)
```

```r
prep_window(c(1,3))

x <- seq(0,10,0.1)
plot(x, dnorm(x), type = "n", xlab = "", ylab = "Density", main = "X", lwd = 2, ylim = c(0,0.4))
for(i in 1:length(scale_list)) lines(x, dgamma(x, shape = 2, scale = scale_list[i]), type = "l", col = scale_cols[i], lwd = 2)
legend("topright", legend = paste0("scale = ",scale_list), lty = 1, lwd = 2, col = scale_cols)

x <- seq(0,75,0.1)
plot(density(nmax, from = min(x), to = max(x)), type = "n", lwd = 2, main = "max (X)", xlab = "", ylim = c(0,0.4))
for(i in 1:length(scale_list)) {
    lines(density(g_max[,i], from = min(x), to = max(x)), lwd = 1, col = scale_cols[i])
    lines(x, devd(x, loc = g_pars["location",i], scale = g_pars["scale",i], shape = g_pars["shape",i]), col = scale_cols[i], lwd = 2)
}

plot(scale_list, g_pars["shape",], col = scale_cols, pch = 4, lwd = 2, ylim = c(-0.1, 0.1), 
     xlab = "Scale parameter of X", ylab = "Shape of max (X) GEV", main = "GEV shape parameter vs gamma scale parameter")
abline(h = 0, lty = 2)
```

```r
prep_window(c(1,3))

plot(scale_list, g_pars["location",], col = scale_cols, pch = 4, lwd = 2, 
     xlab = "Scale parameter of X", ylab = "Location of max (X) GEV", main = "GEV location vs gamma scale parameter")
abline(line(scale_list, g_pars["location",]), lty = 2)

plot(scale_list, g_pars["scale",], col = scale_cols, pch = 4, lwd = 2, 
     xlab = "Scale parameter of X", ylab = "Scale of max (X) GEV", main = "GEV scale vs gamma scale parameter")
abline(line(scale_list, g_pars["scale",]), lty = 2)

plot(scale_list, g_pars["shape",], col = scale_cols, pch = 4, lwd = 2, ylim = c(-0.1,0.1),
     xlab = "Scale parameter of X", ylab = "Shape of max (X) GEV", main = "GEV shape vs gamma scale parameter")
abline(line(scale_list, g_pars["shape",]), lty = 2)
```

### **Changes in both scale & shape parameters**

```r
shape_list <- seq(1,50,3); scale_list <- c(0.5,2,3,4,5); scale_cols <- c("midnightblue", "blue", "cornflowerblue", "skyblue", "aquamarine", "green2")

gmax_pars <- abind::abind(sapply(scale_list, function(theta) {
    sapply(shape_list, function(k) {
        samp <- sapply(1:10000, function(s) max(rgamma(1000, shape = k, scale = theta)))
        fevd(samp)$results$par
    })
}, simplify = F), rev.along = 0)
```

```r
prep_window(c(1,3))
matplot(shape_list, gmax_pars["location",,], pch = 4, lwd = 2, col = scale_cols,
        main = "GEV location parameter", xlab = "Gamma shape parameter", ylab = "GEV location parameter")

matplot(shape_list, gmax_pars["scale",,], pch = 4, lwd = 2, col = scale_cols,
        main = "GEV scale parameter", xlab = "Gamma shape parameter", ylab = "GEV scale parameter")

matplot(shape_list, gmax_pars["shape",,], pch = 4, lwd = 2, col = scale_cols, ylim = c(-0.1, 0.1),
        main = "GEV shape parameter", xlab = "Gamma shape parameter", ylab = "GEV shape parameter")
abline(h = 0, lty = 2)
legend("topright", legend = scale_list, col = scale_cols, lty = 1, lwd = 2, title = "Gamma scale parameter")
```
