```r
library(extRemes)
```

```r
# simulate from known GEV
nsamp <- 5000
return_levels <- 2:200
shape_parameters <- c(-0.25, 0, 0.25)

xc <- sapply(shape_parameters, function(xi) {
    x <- revd(nsamp, shape = xi)
    mdl <- fevd(x)
    1/sapply(return.level(mdl, return_levels), function(th) sum(x >= th)/nsamp)
})
```

```r
matplot(return_levels, xc, col = c("blue", "black", "red"), pch = 20, xlab = "Expected exceedances", ylab = "Observed exceedances")
abline(0,1, lty = 2)
legend("topleft", paste0("xi = ", shape_parameters), col = c("red", "black", "blue"), pch = 20)
```
