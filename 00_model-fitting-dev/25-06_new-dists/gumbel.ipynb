```r
df <- merge(read.table("gmst.txt", col.names = c("year", "gmst")), read.csv("cpc_mam4.csv"))
```

## Test latest

```r
mdl <- fit_ns("gumbel", "fixeddisp", data = df, covnm = "gmst", varnm = "precip")
mdl_x <- fit_ns("gumbel", "fixeddisp", data = df[df$year < 2024,], covnm = "gmst", varnm = "precip", ev = df$precip[df$year == 2024])


cov_2024 <- df[df$year == 2024,c("gmst"),drop = F]
cov_pi <- cov_2024 - 1.3
```

```r
prep_window(c(1,3))

plot_trend(mdl_x)
plot_covtrend(mdl_x, "gmst")
plot_returnlevels(mdl_x, cov_f = cov_2024, cov_cf = cov_pi)
```
